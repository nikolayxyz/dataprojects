### Шаг 2. Предобработка данных

```python
data.columns[data.isnull().any()]
```

В данных столбцах есть пропущенные значения. Рассмотрим подробнее каждый из столбцов.

```python
data.isnull().sum(axis=0).sort_values().tail(14)
```

<br>В locality_name всего 49 пропущенных - от них мы можем избавиться, однако в некоторых строках есть информация о расстоянии до аэропорта. Т.к. данные только по Санкт-Петербургу и близлежащих городах, мы можем заполнить пропуски в locality_name исходя из информации о расстояния до аэропорта.
<br>Возможная причина появления пропуска - забывчивость пользователя/сбой приложения.

```python
list_of_cities_for_airport=data.groupby([pd.cut(data['airports_nearest'],np.arange(0,85001,1000))])['locality_name']
list_of_cities_for_airport = list_of_cities_for_airport.apply(pd.Series.mode)
#список для заполнения
 
def fillnawithmode(row):
    locality= row['locality_name']
    nearest_airport = row['airports_nearest']
   
    for i in range(0, len(list_of_cities_for_airport)):
        if nearest_airport in list_of_cities_for_airport.index[i][0]:
            return list_of_cities_for_airport[i]
    return locality    
```

```python
data.loc[data['locality_name'].isnull(),'locality_name']=data.apply(fillnawithmode, axis = 1)
```

```python
data['locality_name'].isnull().sum() 
```

```python
data[data['floors_total'].isnull()].head()
```

Пропущенных значений немного - можно и удалить, а можно заменить на какое-нибудь значение и просто не учитывать их при дальнейшем анализе.
<br>Заменим пропуски в этом столбце на нули - такого значения не может быть у реального объекта.

```python
data['floors_total']=data['floors_total'].fillna(0)
```

```python
data.query('floors_total>30')['floors_total'].value_counts()
```

Мы также видим некоторые единичные значения, кажущиеся нереальными, однако на данном этапе решаем их не удалять.


В датасете есть пропущенные значения в living_area и kitchen_area.
<br>Мы вычислим среднее соотношение жилой площади/площади кухни к общей площади для квартир с одинаковым количеством комнат и исходя из этого соотношения заполним пропуски.

```python
list_fill_living=data.groupby('rooms')['living_area'].mean()/data.groupby('rooms')['total_area'].mean()
#список для заполнения пропусков в living_area
```

```python
list_fill_kitchen=data.groupby('rooms')['kitchen_area'].mean()/data.groupby('rooms')['total_area'].mean()
#список для заполнения пропусков в kitchen_area
```

```python
def fill_living(row):
    total=row['total_area']
    rooms=row['rooms']
    living=row['living_area']
    return round(list_fill_living[rooms]*total,2)
 
```

```python
def fill_kitchen(row):
    total=row['total_area']
    rooms=row['rooms']
       
    return round(list_fill_kitchen[rooms]*total,2)
```

```python
data.loc[data['living_area'].isnull(),'living_area']=data.apply(fill_living, axis = 1)
```

```python
data.loc[data['kitchen_area'].isnull(),'kitchen_area']=data.apply(fill_kitchen, axis = 1)
```

<br>Заполним часть пропусков медианой в зависимости от количества этажей в доме, предполагая, что в домах одинаковой этажности будут одинаковые потолки.

```python
data['ceiling_height']=data.groupby('floors_total')['ceiling_height'].transform(lambda x: x.fillna(x.mean()))
```

### Шаг 3. Посчитайте и добавьте в таблицу

Добавим в таблицу цену за квадратный метр.<br>Для этого напишем функцию и присвоим результат ее выполнения в наш новый столбец.

```python
def addpricepersqm(row):
    total_area=row['total_area']
    price=row['last_price']
    return round(price/total_area,2)
```

```python
data['price_per_square_m']=data.apply(addpricepersqm,axis=1)
```

Добавим столбцы с днем недели, месяцем и годом добавления объявления. Для этого воспользуемся функциями для работы с датами.

```python
data['weekday_exposition']=data['first_day_exposition'].dt.day_name()
data['month_exposition']=data['first_day_exposition'].dt.month_name()
data['year_exposition']=pd.DatetimeIndex(data['first_day_exposition']).year
```

В столбцах weekday_exposition и month_exposition хранятся названия дней недели и названия месяцев размещения объявлений.<br>В случае группировки данные будут сортироваться в лексикографическом порядке, что нам не подходит, поэтому заменим их тип данных на упорядоченную категорию.

```python
from pandas.api.types import CategoricalDtype
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
cat_type = CategoricalDtype(categories=cats, ordered=True)
data['weekday_exposition'] = data['weekday_exposition'].astype(cat_type)

cats_2=['January','February','March','April','May','June','July','August','September','October','November','December']
cat_type_2=CategoricalDtype(categories=cats_2, ordered=True)
data['month_exposition']=data['month_exposition'].astype(cat_type_2)
```

Добавим столбцы с названием этажа, отношением жилой площади к общей и площади кухни к общей.<br>Для этого также напишем функции и присвоим их в новые столбцы.

```python
def add_floor_label(row):
    floor=row['floor']
    floors_total=row['floors_total']
    if floor==1:
        return 'Первый'
    if floor==floors_total:
        return 'Последний'
    else:
        return 'Другой'
```

```python
data['floor_label']=data.apply(add_floor_label,axis=1)
```

```python
def add_living_to_total(row):
    living_area=row['living_area']
    total_area=row['total_area']
    return round(living_area/total_area,2)
def add_kitchen_to_total(row):
    kitchen_area=row['kitchen_area']
    total_area=row['total_area']
    return round(kitchen_area/total_area,2)
```

```python
data['living_to_total']=data.apply(add_living_to_total,axis=1)
data['kitchen_to_total']=data.apply(add_kitchen_to_total,axis=1)
```

Также добавим несколько столбцов, которые потребуются нам в дальнейшем.

```python
data['year-month']=data['first_day_exposition'].dt.strftime('%Y-%m')
data['km_to_center']=round(data['cityCenters_nearest']/1000)
#я добавил эти столбцы в данном разделе, чтобы при срезе данных они уже были
```
